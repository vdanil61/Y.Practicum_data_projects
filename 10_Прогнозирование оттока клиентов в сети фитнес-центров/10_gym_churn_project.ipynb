# Прогнозирование оттока клиентов в сети фитнес-центров

Сеть фитнес-центров разрабатывает стратегию взаимодействия с клиентами на основе аналитических данных.
Распространённая проблема фитнес-клубов и других сервисов — отток клиентов. Как понять, что клиент перестал ходить в центр? Можно записать в отток тех, кто попросил закрыть договор или удалил аккаунт. Однако клиенты не всегда уходят демонстративно: чаще перестают пользоваться сервисом тихо.
Индикаторы оттока зависят от специфики отрасли. Когда пользователь редко, но стабильно закупается в интернет-магазине — не похоже, что он «отвалился». А вот если две недели не заходит на канал с ежедневно обновляемым контентом, скорее всего подписчик заскучал и покинул магазин.
Для фитнес-центра можно считать, что клиент попал в отток, если за последний месяц ни разу не посетил спортзал. Конечно, не исключено, что он уехал на долгое времся и по приезде обязательно продолжит ходить на фитнес, однако чаще бывает наоборот. Если клиент немного походил в спортзал, а потом пропал — скорее всего, он не вернётся.
Чтобы бороться с оттоком, отдел по работе с клиентами фитнес-центра перевёл в электронный вид множество клиентских анкет. Задача проекта — провести анализ и подготовить план действий по удержанию клиентов, а именно:
- Научиться прогнозировать вероятность оттока (на уровне следующего месяца) для каждого клиента;
- Сформировать типичные портреты клиентов: выделить несколько наиболее ярких групп и охарактеризовать их основные свойства;
- Проанализировать основные признаки, наиболее сильно влияющие на отток;
- формулировать основные выводы и разработать рекомендации по повышению качества работы с клиентами:
  - 1) выделить целевые группы клиентов;
  - 2) предложить меры по снижению оттока;
  - 3) определить другие особенности взаимодействия с клиентами.

### Анализ распрелений признаков

**Вопрос 3 - Каким образом распределены значения каждого признака в группах по отдельности?**

Для построения графиков немного поработаем с исходным датасетом:
- Разделим на 2 группы (произошел/не произошел отток)
- Уберем значение целевой переменной из датасетов (нужно для построения графиков циклом)

Затем построим графики циклом с перебором названий столбцов. Подойдет тип `sns.distplot` с наложением и разными цветами - так мы явно увидим превышение той или иной группы

```python
# выделение 2-х датасетов по группам
data_gym_0 = data_gym.query('churn == 0').drop('churn', axis = 1)
data_gym_1 = data_gym.query('churn == 1').drop('churn', axis = 1)

# построение графиков
for feature in data_gym_0.columns:
    plt.figure(figsize = (6, 6))
    plt.title('Распределение признака: ' + feature)
    sns.distplot(data_gym_0[feature], bins = 10, label = 'Оставшиеся', color = 'green')
    sns.distplot(data_gym_1[feature], bins = 10, label = 'Ушедшие', color = 'red')
    plt.legend()
    plt.show()
```

Судя по графикам распределений, подтверждаются выводы из предыдущего раздела.

Чаще уходят такие клиенты, которые:
- Живут далеко от центра (в другом районе)
- Не имеют корпоративной скидки или друзей в центре
- Покупают абонементы на короткий срок - чаще всего не более 2 месяцев
- Тренируются одни
- Тренирутся не чаще 2 раз в неделю
- Средний возраст меньше 30 лет

На количественные признаки не обращаем внимания - если клиенты уходят через короткий срок и берут абонемент на пробу, то вполне естесственно, что они тренируются меньше оставшихся клиентов и в меньшем объеме пользуются доп услугами центра (меньше средний чек).

В целом распределения похожи на правду, но наблюдается несколько выбросов у признаков `contract_period`, `month_to_end_contract` и `lifetime` в большую сторону - то есть клиенты уходят после довольно длительного периода. В реальности такое тоже может быть и связано с изменением жизненных обстоятельств - присутствие таких значений допустимо и не сильно повлияет на анализ.

### Матрица корреляций признаков

**Вопрос 4 - Насколько сильно коррелируют признаки между собой - возможна ли мультиколлинеарность?**

Построим матрицу корреляций и посмотрим на связанность признаков между собой попарно.

```python
# построение матрицы корреляций
plt.figure(figsize = (15, 15))
sns.heatmap(data_gym.corr(),
            cmap = 'Blues',
            annot = True,
            square = True,
            annot_kws = {"size": 8},
            fmt = '.2g',
            vmin = -1, vmax = 1)
plt.show()
```

Судя по матрице корреляций, большинство признаков слабо связаны между собой. Но некоторые пары признаков вызывают много внимания:
- Пара `month_to_end_contract` и `contract_period` - корреляция равна 0.97.
- Пара `avg_class_frequency_current_month` и `avg_class_frequency_total` - корреляция равна 0.95.

## Обучение моделей прогнозирования оттока клиентов

### Выделение выборок, стандартизация и обучение моделей

Перед нами стоит задача прогнозирования оттока клиентов. Отток - либо может состояться, либо не состоится. Соответственно, решаемая задача на языке ML - **бинарная классификаця**.

Для обучения используем выборку из 4 тыс. наблюдений и 11 признаков - `data_gym_clean`

Выберем несколько типов моделей для обучения:
- Логистическая регрессия - `LogisticRegression`
- Случайный лес - `RandomForestClassifier`
- Градиентный бустинг - `GradientBoostingClassifier` (дополнительно)

Первым делом разобъем признаки и целевую переменную (отток клиента), выделим имеющиеся данные на 2 выборки - тестовую и валидационную в соотношении 80/20. Затем стандартизируем данные

```python
# разделение признаков и целевой переменной
X = data_gym_clean.drop('churn', axis = 1)
y = data_gym_clean['churn']

# выделение тестовой и валидационной выборок
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0, test_size=0.2)

# создание объекта класса StandardScaler и применение к обучающей выборке
scaler = StandardScaler()
scaler.fit(X_train)

# преобразуем тестовую и валидационную выборки
X_train_st = scaler.transform(X_train)
X_test_st = scaler.transform(X_test)
```

Датасет разделили, выборки признаков стандартизировали - все вводные для обучения получены. Можем приступать к обучению моделей - попробуем создать функцию и пройтись циклом по списку моделей (сформируем отдельно).

Оценивать модели предполагается по 4 метрикам:
- 1) **Accuracy** - доля правильных ответов алгоритма
- 2) **Precision** - точность модели. Доля правильных ответов только среди целевого класса. То есть модель реже сработает и не подаст знак на того клиента, который на самом деле может и не уйти
- 3) **Recall** - полнота модели. Показывает, сколько реальных фактов ухода клиентов удалось предсказать верно
- 4) **F1-мера** - сводная метрика, учитывающая баланс между Precision и Recall

```python
# формирование списка моделлей
models = [LogisticRegression(solver = 'liblinear', random_state = 0),
          RandomForestClassifier(n_estimators = 100, random_state = 0),
          GradientBoostingClassifier(n_estimators = 100, random_state = 0)
         ]

# функция для обучения моделей и вывода метрик по каждой
def make_predictions(model, X_train, y_train, X_test, y_test):
    
    # задание и обучение моделей
    model = model
    model.fit(X_train, y_train)
    
    # прогнозирование целевой переменной и вероятности
    predictions = model.predict(X_test)
    probabilities = model.predict_proba(X_test)[:,1]
    
    # вывод метрик
    print('Accuracy: {:.2f}'.format(accuracy_score(y_test, predictions)))
    print('Precision: {:.2f}'.format(precision_score(y_test, predictions)))
    print('Recall: {:.2f}'.format(recall_score(y_test, predictions)))
    print('F1: {:.2f}'.format(f1_score(y_test, predictions)))
```

Функцию для обучения моделей получили. Теперь пройдем циклом и сравним модели между собой.

```python
# задание цикла с обучением моделей и выводом метрик
for model in models:
    print('Модель:', str(model).split('(')[0]) # хитрый способ убрать параметры модели при выводе
    make_predictions(model, X_train_st, y_train, X_test_st, y_test)
    print('')
```

Судя по метрикам, самой лучшей оказалась модель, обученная градиентным бустингом. При почти одинаковой метрике Accuracy с моделями логистической регрессии и случайного леса, значения Precision, Recall оказались очень близки друг к другу и обладают минимальной разницей (они связаны между собой). Также модель градиентного бустинга обладает наибольшим значением F-меры.

**Для дальнейшего использования выбираем модель градиентного бустинга**

## Кластеризация клиентов

На этом шаге попробуем сегментировать клиентов, определить наиболее склонных к оттоку. **На языке ML - это задача кластеризации, обучение "без учителя"**. Модели важно уловить схожесть клиентов между собой и разделить общий поток клиентов на несколького сегментов - это позволит точнее предлагать акции/скидки или определенный набор доп услуг для удержания клиентов.

Первым делом стандартизируем общий набор данных и построим дендрограмму с разделением по кластерам. Используем датасет `data_gym` - при кластеризации взаимные корреляции признаков не так важны

### Построение дендрограммы и определение количества кластеров

Первым делом стандартизируем общий набор данных и построим дендрограмму с разделением по кластерам

```python
# создание объекта класса StandardScaler и применение к общей выборке, с преобразованием данных
scaler = StandardScaler()

X_st = scaler.fit_transform(X)
```

Общий набор данных стандартизировали, строим дендрограмму

```python
# рассчет матрицы расстояний
linked = linkage(X_st, method = 'ward')

# построение дендрограммы
plt.figure(figsize = (15, 15))
dendrogram(linked, orientation = 'top', truncate_mode = 'lastp')
plt.title('Дендрограмма по клиентам сети фитнес-центров')
plt.xlabel('Размер кластера')
plt.ylabel('Расстояние')
plt.show()
```

Дендрограмму получили - можем выделить 4 отдельных кластера. В соответствии с условиями проекта, выделяем 5 отдельных кластеров для будущего обученмя модели.

Запомнили `n = 5`, приступаем к обучению модели кластеризации "без учителя"

### Обучение модели кластеризации

Количество кластеров предварительно определили до обучения модели, возможно, это позволит обучить модель быстрее. Алгоритм `KMeans` построен на предположении, что количество кластеров известно заранее.

```python
# задание модели кластеризации
km_model = KMeans(n_clusters = 5, random_state = 0)

# обучение модели и получение кластеров по каждому клиенту
labels = km_model.fit_predict(X_st)

# занесение номера кластера в датасет новым столбцом
data_gym['km_cluster'] = labels
```

Модель кластеризации обучили. Выведем, ради интереса, количество клиентов в каждом кластере

```python
# вывод количества клиентов по кластерам
data_gym.groupby('km_cluster')['churn'].count()
```

Обучение сработало - получено отнесение каждого клиента к определенному кластеру

### Общие выводы

В рамках проекта произведен анализ причин оттока клиентов из сети фитнес-центров, градиентным бустингом обучена ML-модель, способная предсказывать вероятность оттока клиентов, и также с применением ML произведена кластеризация клиентов по 5 кластерам.

Лучшая модель по метрикам - модель, обученная градиентным бустингом. Модель имеет значения метрик:
- Accuracy: 0.91
- Precision: 0.81
- Recall: 0.80
- F1: 0.81

Определены ключевые признаки, влияющие на отток клиентов, на которые стоит опираться при проработке активностей по удержанию:
- `lifetime`
- `month_to_end_contract`
- `age`
- `avg_class_frequency_total`

Дополнительно была обучена ML-модель кластеризации, по которой выделено 5 основных кластеров клиентов. Наиболее важные кластеры:
- **2 - самый надежный кластер** с точки зрения удержания клиентов. Основная аудитория - тренируется чаще всего (2-3 раза в неделю), приобретает длительные абонементы (год), живут или работают рядом с фитнес-центрами
- **0 - самый показательный кластер** с точки зрения признаков и 2-ой по качеству удержания клиентов. Аудитория кластера в подавляющем большинстве живет или работает рядом с фитнес-центрами, все пришли по наводкам друзей и затем пользуются скидками от работодателя, имеет самый высокий показатель по факту посещения групповых занятий, абонементы покупаются чаще всего на длительный срок (год)
- **3 - самый ненадежный кластер** с точки зрения удержания клиентов. Основная аудитория - живет или работает недалеко от центров, реже всего тренируется и покупает абонементы не более,чем на полгода. Скорее всего, клиенты прихоят попробовать занятия, но в силу тех или иных обстоятельств отказываются от тренировок - либо не смогли выстроить график, либо нашли вариант выгоднее

### Рекомендации сети фитнес-центров

**Основная рекомендация: сети стоит сконцентироваться на удержании новых клиентов на самых первых этапах.**

*Судя по кластеру 3, клиенты чаще всего уходили в самом начале, тренировались одни и редко*

*В кластере 0, напротив, все с нетворкингом очень хорошо - на такую модель можно ориентироваться*

Варианты удержания клиентов на первых этапах:

- 1) **Предложить welcome-бонус всем новым клиентам**

В качестве бонуса могут быть сертификаты на сопутствующие услуги - массаж, спа после тренировок или дополнительно час в бассейне (тут смотря какой центр и есть ли бассейн). Это поможет задобрить клиента и дать ему ощущение заботы. Второй вариант - какой-нибудь небольшой "входной" аксессуар: бутылка для воды. Третий - устроить короткую консультацию с инструктором, предложить сформировать базовые принципы тренировок и программу обязательных упражнений/разминки, исходя из цели начала тренировок.

- 2) **Внедрение геймификации, челленджей среди новичков**

Новичками волне могут считаться клиенты в первые 1-2 месяца тренировок. Для стимулирования активностей и более частых тренировок можно ввести своеобразные соревнования/рейтинги, и каждый месяц/пару недель выдавать призы лидерам - что-то покруче небольшого аксессуара. Например, футболку с лого сети центров, дополнительную скидку при будущем продлении абонемента. Второй эффект (побочный, но очень важный для лайфтайма) - клиент под азартом, незаметно для себя, может выстроить удобный график тренировок, они со временем войдут в привычку.

- 3) **Стимулирование новичков на групповые занятия**

Также возможен вариант, что новички могут стесняться своей фигуры, техники со снарядами/тренажерами, да и просто быть замкнутыми. Как вариант "раскрепощения" - сагитировать на групповые тренировки для новичков. Такой вариант позволит психологически воздействовать на клиента и позволит ему понять, что он не один новичок - со временем как минимум раскрепостится, а может даже и найдет собеседников/друзей, с которыми вместе будет ходить на тренировки.
